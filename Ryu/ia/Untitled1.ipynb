```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Chargement des donnes (remplacez cela par vos propres donnes)
# Exemple avec des donnes alatoires pour illustrer le fonctionnement
np.random.seed(42)
data = np.random.rand(100, 4)

# Cration du modle K-means avec 3 clusters
kmeans = KMeans(n_clusters=4, random_state=42)

# Entranement du modle
kmeans.fit(data)

# Prdictions des clusters pour chaque observation
labels = kmeans.labels_

# Coordonnes des centrodes
centroids = kmeans.cluster_centers_

# Visualisation des clusters
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', marker='o', edgecolors='black')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('Clustering avec K-means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```
